```python
btc_price= pd.read_csv("C:/Dissertation/ML Codes Final/ML Output/bitcoin_op.csv")
```

```python
btc_price['PercentageDifference'] = (btc_price['Predicted_Bitcoin_Price'] - btc_price['Predicted_Bitcoin_Price'].shift(1)) / abs(btc_price['Predicted_Bitcoin_Price']) * 100
btc_price['PercentageDifference'].iloc[0] = float('NaN')
data= btc_price['Predicted_Bitcoin_Price']
sample_size = len(data)
standard_deviation = np.std(data)
standard_error = standard_deviation / np.sqrt(sample_size)

z = stats.norm.ppf(1 - (1 - 0.95) / 2)

may_31st_data = btc_price[btc_price['Date'] == '2023-05-30']

lower_bound_b = round(may_31st_data['Predicted_Bitcoin_Price'].iloc[0] - z * standard_error,1)

upper_bound_b = round(may_31st_data['Predicted_Bitcoin_Price'].iloc[0] + z * standard_error,1)


btc_direction = may_31st_data['Predicted_Direction'].iloc[0]
percentage_difference = may_31st_data['PercentageDifference'].iloc[0]
formatted_value_btc = "{:.2f}%".format(percentage_difference)
predicted_price_btc= may_31st_data['Predicted_Bitcoin_Price'].iloc[0]
date= may_31st_data['Date'].iloc[0]
```

```python
import os
import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText


# In[6]:


# Email configuration
SMTP_SERVER = 'smtp.gmail.com'
SMTP_PORT = 587
SENDER_EMAIL = 'felixjosekannai@gmail.com'
RECIPIENT_EMAIL = 'felixjosekannai@gmail.com'


# In[7]:


def send_email(subject, body):
    try:
        # Connect to the SMTP server
        server = smtplib.SMTP(SMTP_SERVER, SMTP_PORT)
        server.starttls()

        # Log in to your email account using environment variables
        email_password = ''
        server.login(SENDER_EMAIL, email_password)

        # Create a multipart message container
        msg = MIMEMultipart()
        msg['From'] = SENDER_EMAIL
        msg['To'] = RECIPIENT_EMAIL
        msg['Subject'] = subject

        # Add message body
        msg.attach(MIMEText(body, 'plain'))

        # Send the email
        server.send_message(msg)

        # Disconnect from the server
        server.quit()
        print('Email sent successfully!')
    except Exception as e:
        print('Error sending email:', str(e))


# In[8]:


# Example usage
subject = 'Daily Cryptocurrency Analysis and Prediction Report for 2023-05-30'


body = f'''
Dear [Manager's Name],

I hope this email finds you well. As part of our ongoing project to predict cryptocurrency price movements, I have generated the following outputs:

{output}



Please let me know if you need any further information.

Best regards,
[Your Name]
'''


# In[9]:


send_email(subject, body)

```
